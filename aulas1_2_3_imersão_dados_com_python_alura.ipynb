```python

# Dicionário de renomeação
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Aplicando renomeação
df.rename(columns=novos_nomes, inplace=True)

# Verificando resultado
df.head()
```

```python
df["senioridade"].value_counts()
```

```python
df["contrato"].value_counts()
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
senioridade = {
    "SE": "Senior",
    "MI": "Pleno",
    "EN": "Junior",
    "EX": "Executivo"
}
df["senioridade"] = df["senioridade"].replace(senioridade)
df["senioridade"].value_counts()
```

```python
contrato_traduzido = {
    "FT": "Tempo Integral",
    "CT": "Contratado",
    "PT": "Meio Período",
    "FL": "Freelancer"
}
df["contrato"] = df["contrato"].replace(contrato_traduzido)
df["contrato"].value_counts()
```

```python
tamanho_empresa_traduzido = {
    "M": "Médio",
    "L": "Grande",
    "S": "Pequeno"
}
df["tamanho_empresa"] = df["tamanho_empresa"].replace(tamanho_empresa_traduzido)
df["tamanho_empresa"].value_counts()
```

```python
remoto_traduzido = {
    0: "Presencial",
    50: "Híbrido",
    100: "Remoto"
}
df["remoto"] = df["remoto"].replace(remoto_traduzido)
df["remoto"].value_counts()
```

```python
df.describe(include="object")
```

```python
import numpy as np
df_salarios = pd.DataFrame({
    "nome": ["Ana", "joão", "bruno", "carlos", "sofia"],
    "salario": [1000, np.nan, 3000, np.nan, 1000000]
})
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame ({
    "dia" : ["segunda", "terça", "Quarta", "quinta", "sexta"],
    "temperatura": [30, np.nan, np.nan, 28, 27]
})
df_temperaturas["preenchido_ffill"] = df_temperaturas["temperatura"].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame ({
    "dia" : ["segunda", "terça", "Quarta", "quinta", "sexta"],
    "temperatura": [30, np.nan, np.nan, 28, 27]
})
df_temperaturas["preenchido_bfill"] = df_temperaturas["temperatura"].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    "nomes" : ["Ana", "joão", "bruno", "carlos", "sofia"],
    "cidade" : ["Belém", np.nan, "Curitiba", np.nan, "São Paulo"]
})
df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("não informado")
df_cidades
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="senioridade", y="usd", data=df_limpo)
plt.title("Salário Médio por Nível de Senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salário médio anual(USD)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="senioridade", y="usd", data=df_limpo, order=ordem)
plt.title("Salário Médio por Nível de Senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salário médio anual(USD)")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_limpo["usd"], bins=50, kde=True)
plt.title("Distribuição dos Salários Anuais")
plt.xlabel("Salário Anual(USD)")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df_limpo["usd"])
plt.title("Boxplot Salários Anuais")
plt.xlabel("Salário Anual(USD)")
plt.show()
```

```python
ordem_senioridade = ["Junior", "Pleno", "Senior", "Executivo"]

plt.figure(figsize=(8, 5))
sns.boxplot(x="senioridade", y="usd", data=df_limpo, order=ordem_senioridade)
plt.title("Boxplot Salários Anuais por Nível de Senioridade")
plt.xlabel("Salário em usd")
plt.show()
```

```python
ordem_senioridade = ["Junior", "Pleno", "Senior", "Executivo"]

plt.figure(figsize=(8, 5))
sns.boxplot(x="senioridade", y="usd", data=df_limpo, order=ordem_senioridade, palette="Set2", hue="senioridade")
plt.title("Boxplot Salários Anuais por Nível de Senioridade")
plt.xlabel("Salário em usd")
plt.show()
```

```python
import plotly.express as px
```

```python
df_media_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reset_index()
df_media_senioridade = df_media_senioridade.sort_values('usd', ascending=False)

fig = px.bar(df_media_senioridade,
             x='senioridade',
             y='usd',
             title='Salário Médio Anual por Nível de Senioridade',
             labels={'senioridade': 'Nível de Senioridade', 'usd': 'Salário Médio Anual (USD)'},
             color='senioridade', # Optional: color bars by seniority
             template='plotly_white') # Optional: use a clean template

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção do tipo de trabalho',
)
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção do tipo de trabalho',
             hole=0.5
)
fig.update_traces(textinfo="percent+label")
fig.show()
```
